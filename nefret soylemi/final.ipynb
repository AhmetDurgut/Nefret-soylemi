```python
# Google Drive'Ä± baÄŸlayÄ±n
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Dosya yollarÄ±
train_file = '/content/drive/MyDrive/oÌˆdev/offenseval-tr-training-v1.tsv'
test_file = '/content/drive/MyDrive/oÌˆdev/offenseval-tr-testset-v1.tsv'

# EÄŸitim verisini yÃ¼kleyin
train_df = pd.read_csv(train_file, sep='\t')

# Ä°lk birkaÃ§ satÄ±rÄ± gÃ¶rÃ¼ntÃ¼leyin
print("EÄŸitim verisi ilk birkaÃ§ satÄ±r:")
print(train_df.head())

# Metin ve etiket sÃ¼tunlarÄ±nÄ± belirleyin
X_train = train_df['tweet']  # Metin sÃ¼tununuzun adÄ±nÄ± buraya yazÄ±n
y_train = train_df['subtask_a']  # Etiket sÃ¼tununuzun adÄ±nÄ± buraya yazÄ±n

# TF-IDF vektÃ¶rleÅŸtirme
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)

# Modeli oluÅŸturun ve eÄŸitin
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Test verisini yÃ¼kleyin
test_df = pd.read_csv(test_file, sep='\t')

# Ä°lk birkaÃ§ satÄ±rÄ± gÃ¶rÃ¼ntÃ¼leyin
print("Test verisi ilk birkaÃ§ satÄ±r:")
print(test_df.head())

# Test verisindeki metin sÃ¼tununu belirleyin
X_test = test_df['tweet']  # Metin sÃ¼tununuzun adÄ±nÄ± buraya yazÄ±n

# Test verisini TF-IDF ile dÃ¶nÃ¼ÅŸtÃ¼rÃ¼n
X_test_tfidf = vectorizer.transform(X_test)

# Tahminlerde bulunun
test_df['predicted_label'] = model.predict(X_test_tfidf)

# Nefret sÃ¶ylemi iÃ§eren kelimeler ve ifadeler
hate_speech_keywords = [
    'geri zekalÄ±', 'aptal', 'pis', 'ÅŸerefsiz', 'hain', 'Ä±rkÃ§Ä±', 'cinsiyetÃ§i', 'dini', 'koyduÄŸum', 'engelli',
    'kafir', 'mÃ¼nafÄ±k', 'terÃ¶rist', 'yobaz', 'faÅŸist', 'Ä±rz dÃ¼ÅŸmanÄ±', 'tecavÃ¼zcÃ¼', 'ahlaksÄ±z', 'pislik', 'sapÄ±k', 'piÃ§', 'orospu', 'siktir'
]

# Nefret sÃ¶ylemi iÃ§eren tweetleri filtreleyin
pattern = '|'.join(hate_speech_keywords)
hate_speech_df = test_df[test_df['tweet'].str.contains(pattern, case=False, na=False)]

# Nefret sÃ¶ylemi olarak sÄ±nÄ±flandÄ±rÄ±lan ve nefret sÃ¶ylemi iÃ§eren kelimeleri filtreleyin
final_hate_speech_df = test_df[(test_df['predicted_label'] == 'nefretsÃ¶ylemi') | (test_df['tweet'].str.contains(pattern, case=False, na=False))]

# Nefret sÃ¶ylemi olarak sÄ±nÄ±flandÄ±rÄ±lanlarÄ± gÃ¶rÃ¼ntÃ¼leyin
print("Nefret sÃ¶ylemi olarak sÄ±nÄ±flandÄ±rÄ±lanlar ve nefret sÃ¶ylemi iÃ§erenler:")
print(final_hate_speech_df)

# SonuÃ§larÄ± kaydedin
output_file_path = '/content/drive/My Drive/hate_speech_detected.tsv'
try:
    final_hate_speech_df.to_csv(output_file_path, sep='\t', index=False)
    print("Dosya baÅŸarÄ±yla kaydedildi.")
except Exception as e:
    print(f"Dosya kaydedilemedi: {e}")





```
