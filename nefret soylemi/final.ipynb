```python
# Google Drive'ı bağlayın
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Dosya yolları
train_file = '/content/drive/MyDrive/ödev/offenseval-tr-training-v1.tsv'
test_file = '/content/drive/MyDrive/ödev/offenseval-tr-testset-v1.tsv'

# Eğitim verisini yükleyin
train_df = pd.read_csv(train_file, sep='\t')

# İlk birkaç satırı görüntüleyin
print("Eğitim verisi ilk birkaç satır:")
print(train_df.head())

# Metin ve etiket sütunlarını belirleyin
X_train = train_df['tweet']  # Metin sütununuzun adını buraya yazın
y_train = train_df['subtask_a']  # Etiket sütununuzun adını buraya yazın

# TF-IDF vektörleştirme
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)

# Modeli oluşturun ve eğitin
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Test verisini yükleyin
test_df = pd.read_csv(test_file, sep='\t')

# İlk birkaç satırı görüntüleyin
print("Test verisi ilk birkaç satır:")
print(test_df.head())

# Test verisindeki metin sütununu belirleyin
X_test = test_df['tweet']  # Metin sütununuzun adını buraya yazın

# Test verisini TF-IDF ile dönüştürün
X_test_tfidf = vectorizer.transform(X_test)

# Tahminlerde bulunun
test_df['predicted_label'] = model.predict(X_test_tfidf)

# Nefret söylemi içeren kelimeler ve ifadeler
hate_speech_keywords = [
    'geri zekalı', 'aptal', 'pis', 'şerefsiz', 'hain', 'ırkçı', 'cinsiyetçi', 'dini', 'koyduğum', 'engelli',
    'kafir', 'münafık', 'terörist', 'yobaz', 'faşist', 'ırz düşmanı', 'tecavüzcü', 'ahlaksız', 'pislik', 'sapık', 'piç', 'orospu', 'siktir'
]

# Nefret söylemi içeren tweetleri filtreleyin
pattern = '|'.join(hate_speech_keywords)
hate_speech_df = test_df[test_df['tweet'].str.contains(pattern, case=False, na=False)]

# Nefret söylemi olarak sınıflandırılan ve nefret söylemi içeren kelimeleri filtreleyin
final_hate_speech_df = test_df[(test_df['predicted_label'] == 'nefretsöylemi') | (test_df['tweet'].str.contains(pattern, case=False, na=False))]

# Nefret söylemi olarak sınıflandırılanları görüntüleyin
print("Nefret söylemi olarak sınıflandırılanlar ve nefret söylemi içerenler:")
print(final_hate_speech_df)

# Sonuçları kaydedin
output_file_path = '/content/drive/My Drive/hate_speech_detected.tsv'
try:
    final_hate_speech_df.to_csv(output_file_path, sep='\t', index=False)
    print("Dosya başarıyla kaydedildi.")
except Exception as e:
    print(f"Dosya kaydedilemedi: {e}")





```
